```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
#x = x_train_norm
#y =  y_train
k_c = [100, 1000, 10000, 100000, 1000000]
n_score_c = []
# loop through range  values f n_stimators
for x_c in k_c:
    #mpement  RandomForestClassifier with n_estimators
    classifier_score = SVC(C = x_c, gamma = 1, random_state = 0)
    classifier_score.fit(logs_train_norm, labels_train)
    #derive cross_val_score for RandomForestClassifier with n_estimators
    scores_c = cross_val_score(estimator = classifier_score, X = logs_train_norm, y = labels_train, cv = 10)
    #add man of scores for n_estmators to n_score list
    n_score_c.append(scores_c.mean())
print( n_score_c)
print(scores_c.mean())
```

```python
#plot the value of n_estimators for RandomForestClassifier (x-axis) versus the cross validated accuracy (y-axis)
x = [100, 1000, 10000, 100000, 1000000]
#y = [0.939, 0.942, 0.941,  0.941, 0.941, 0.943, 0.94, 1., 0.960]
plt.figure(figsize = (6, 4), dpi = 600)
plt.plot(x, n_score_c, linewidth = 3, color = 'green')
plt.xlabel('values of C for svm model')
plt.ylabel('cross validated score')
plt.xscale('log')
#plt.ylim(0.939, 0.943)         
plt.savefig('desktop/modified_b_svm_C_values_used.svg', format = 'svg', dpi = 600)
plt.show()
```

```python
#x = x_train_norm
#y =  y_train
k_g = [1, 10, 100]
n_score_g = []
# loop through range  values f n_stimators
for x_g in k_g:
    #mpement  RandomForestClassifier with n_estimators
    classifier_score = SVC(C = 1000, gamma = x_g, random_state = 0)
    classifier_score.fit(logs_train_norm, labels_train)
    #derive cross_val_score for RandomForestClassifier with n_estimators
    scores_g = cross_val_score(estimator = classifier_score, X = logs_train_norm, y = labels_train, cv = 10)
    #add man of scores for n_estmators to n_score list
    n_score_g.append(scores_g.mean())
print( n_score_g)
print(scores_g.mean())
```
