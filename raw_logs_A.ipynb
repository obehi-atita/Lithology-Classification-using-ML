```python
#defining function to plot borehole log
def well_plot(dataset):
    import matplotlib as mpl
    import matplotlib.colors
    from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,AutoMinorLocator)
    from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
    mpl.rcParams.update(mpl.rcParamsDefault)
    import matplotlib.pyplot as plt
    %matplotlib inline
    import seaborn as sns
    #dataset = dataset.sort_values(by = 'Depth_vp')
    #dataset = elastic_prop.sort_values(by = 'Depth_vp')
   # dataset = elastic_prop['Depth_vp'].values
   # cmap = mpl.colors.ListedColormap(cluster_colors[0:len(cluster_colors)], 'indexed')
    min_depth_ray_a = dataset.depth_ray.min()
    max_depth_ray_a = dataset.depth_ray.max()
    min_depth_density_a = dataset.depth_density.min()
    max_depth_density_a = dataset.depth_density.max()
    min_depth_vs_a = dataset.Depth_vs.min()
    max_depth_vs_a = dataset.Depth_vs.max()
    min_depth_vp_a =dataset.Depth_vp.min()
    max_depth_vp_a = dataset.Depth_vp.max()
   # cluster = np.repeat(dataset[labels].values, 100, 1)
    #cluster = np.expand_dims(dataset[labels].values, 1)
    #print (cluster)
    fig, subplot = plt.subplots(nrows = 1, ncols = 4, figsize = (10, 8))
    #set_share_axes(ax[:, :8], sharey = True)
    #ax.get_xaxis().set_visible(True)
    # ax.get_yaxis().set_visible(True)
    #f.subplots_adjust(right = 5)
    #make tick labels invisible
    #plt.setp(ax2.get_xticklabels(), visible = False),
    #plt.xticks([])
    #ax.xaxis.set_major_locator(ticker,Nulllocator())
    minor_tick = (MultipleLocator(20))#20
    major_tick = (MultipleLocator(100))#100
    major_tick_label = (FormatStrFormatter('%d'))
#minor_tick_2 = np.arange(160, dataset_min_depth, 20)
#major_tick_2 = np.arange(160, dataset_max_depth, 100)
    #fig = plt.figure(figsize = (12, 10)
    #fig = plt.figure() 
   # fig.set_figheight(10)    
    #fig.set_figwidth(14)
    plt.subplots_adjust(wspace = 0.1, hspace = 0.15)                  
    subplot[0].plot(dataset['Gamma ray'], dataset.depth_ray, "-", color = 'yellow')
    subplot[0].set_ylim(0, 840)
    subplot[0].invert_yaxis()
    subplot[0].yaxis.set_minor_locator(minor_tick)
    subplot[0].yaxis.set_major_locator(major_tick)  
    subplot[0].yaxis.set_major_formatter(major_tick_label)
    subplot[0].grid(True, which = "major", linestyle = "-")
    subplot[0].grid(True, which = "minor", linestyle = ":")
    subplot[0].minorticks_on()
    subplot[0].tick_params("y", labelsize = 16, which ='major', length = 6)
    subplot[0].tick_params("y", labelsize = 16, which = 'minor', length = 2) 
    subplot[0].tick_params("x", labelsize = 16, which ='major', length = 6)
    subplot[0].tick_params("x", labelsize = 16, which = 'minor', length = 2)
    plt.setp(subplot[0].get_yticklabels(), visible = False)  
    subplot[1].plot(dataset['density'], dataset.depth_density, "-", color = 'dimgray')
    subplot[1].set_ylim(0, 840)
    subplot[1].invert_yaxis()
    subplot[1].yaxis.set_minor_locator(minor_tick)
    subplot[1].yaxis.set_major_locator(major_tick)  
    subplot[1].yaxis.set_major_formatter(major_tick_label)
    subplot[1].grid(True, which = "major", linestyle = "-")
    subplot[1].grid(True, which = "minor", linestyle = ":")
    subplot[1].minorticks_on()
    subplot[1].tick_params("y", labelsize = 16, which ='major', length = 6)
    subplot[1].tick_params("y", labelsize = 16, which = 'minor', length = 2) 
    subplot[1].tick_params("x", labelsize = 16, which ='major', length = 6)
    subplot[1].tick_params("x", labelsize = 16, which = 'minor', length = 2)
    #subplot[1].tick_params(axis = 'x', labelsize = 6, which ='major', length = 6)
    #subplot[1].tick_params(axis = 'x', labelsize = 6, which = 'minor', length = 2)
    plt.setp(subplot[1].get_yticklabels(), visible = False)   
    subplot[2].plot(dataset["Vp"], dataset.Depth_vp, "-", color = 'cyan')
    subplot[2].set_ylim(0, 840)
    subplot[2].invert_yaxis()
    subplot[2].yaxis.set_minor_locator(minor_tick)
    subplot[2].yaxis.set_major_locator(major_tick)  
    subplot[2].yaxis.set_major_formatter(major_tick_label)
    subplot[2].grid(True, which = "major", linestyle = "-")
    subplot[2].grid(True, which = "minor", linestyle = ":")
    subplot[2].minorticks_on()
    subplot[2].tick_params("y", labelsize = 16, which ='major', length = 6)
    subplot[2].tick_params("y", labelsize = 16, which = 'minor', length = 2) 
    subplot[2].tick_params("x", labelsize = 16, which ='major', length = 6)
    subplot[2].tick_params("x", labelsize = 16, which = 'minor', length = 2)   
    plt.setp(subplot[2].get_yticklabels(), visible = False) 
  
    subplot[3].plot(dataset["Vs"], dataset.Depth_vs, "-", color = 'saddlebrown')
    subplot[3].set_ylim(0, 840)
    subplot[3].invert_yaxis()
    subplot[3].yaxis.set_minor_locator(minor_tick)
    subplot[3].yaxis.set_major_locator(major_tick)  
    subplot[3].yaxis.set_major_formatter(major_tick_label)
    subplot[3].grid(True, which = "major", linestyle = "-")
    subplot[3].grid(True, which = "minor", linestyle = ":")
    subplot[3].minorticks_on()
    subplot[3].tick_params("y", labelsize = 16, which ='major', length = 6)
    subplot[3].tick_params("y", labelsize = 16, which = 'minor', length = 2) 
    subplot[3].tick_params("x", labelsize = 16, which ='major', length = 6)
    subplot[3].tick_params("x", labelsize = 16, which = 'minor', length = 2)   
    plt.setp(subplot[3].get_yticklabels(), visible = False) 
  
```
