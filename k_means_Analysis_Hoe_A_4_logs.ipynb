```python
#defining an empty list of sum of squares
sum_squares_2 = []
```

```python
##looping through range of K values for the optimization of K using elbow method
for k in range(1,11):
    KM = KMeans(n_clusters = k, random_state = 3425)
    KM.fit(prop_main_components_2)
    sum_squares_2.append(KM.inertia_)
```

```python
plt.figure(figsize = (8, 6), dpi = 400)
plt.plot(range(1,11), sum_squares_2, '-o', linewidth = 3)
plt.xlabel('clusters, K', fontsize = 20)
plt.ylabel('sum_squares', fontsize = 20)
plt.xticks(range(1,11))
plt.tick_params(direction = 'out', length = 4, color = 'k')
plt.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.grid(True, which = 'major')
plt.savefig('desktop/k_means_elbow_plot_2_pcas.png', format = 'png', dpi = 400)
plt.show()
```
